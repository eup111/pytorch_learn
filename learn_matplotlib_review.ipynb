```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#高清
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('svg')
```

```python
figure = plt.figure(figsize = (3,3))
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y)
plt.show()
```

```python
figure.savefig(r'D:\matplotlib_review.svg')
```

```python
figure2 = plt.figure(figsize = (3,3))
axes = plt.axes()
axes.plot(x,y)
```

```python
figure3 = plt.figure(figsize = (5,5))
plt.subplot(1,3,1),plt.plot(x,y)
plt.subplot(1,3,2),plt.plot(x,y)
plt.subplot(1,3,3),plt.plot(x,y)
```

```python
figure3 = plt.figure(figsize = (5,5))
plt.subplot(3,1,1),plt.plot(x,y)
plt.subplot(3,1,2),plt.plot(x,y)
plt.subplot(3,1,3),plt.plot(x,y)
```

```python
fig,axes = plt.subplots(3)
axes[0].plot(x,y)
```

```python
x2 = [5,4,3,2,1]
figure4= plt.figure(figsize = (5,5))
plt.plot(x,y,color = 'red',linestyle = '--',linewidth = 2,marker = 'o')
plt.plot(x2,y,color = 'blue',linestyle = ':',linewidth = 3,marker = '^')
plt.scatter(x2,y,color = 'yellow',marker = 's')
plt.show()
```

```python
import numpy as np
x = np.random.randint(0,256,size=(28,28))
print(x.shape)
```

```python
plt.figure(figsize = (5,5))
plt.imshow(x)
plt.colorbar()
plt.show()
```

```python
x = np.random.randn(100)
plt.figure(figsize = (5,5))
plt.hist(x,bins=30,color = 'white',edgecolor = 'black')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.figure(figsize = (5,5))
plt.plot(x,y,label='y=x')
plt.xlim(1,10)
plt.ylim(1,5)
plt.xlabel('xxxx')
plt.ylabel('yyyy')
plt.title('figure')
plt.legend()
plt.show()
```
